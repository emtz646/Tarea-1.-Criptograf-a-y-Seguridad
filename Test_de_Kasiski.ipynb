```python
import math
import re
from itertools import combinations
from collections import Counter

def calcular_cadenas_comunes(texto):
    # Eliminamos caracteres especiales y normalizamos
    texto = texto.replace('Á', 'A').replace('É', 'E').replace('Í', 'I').replace('Ó', 'O').replace('Ú', 'U').replace("\n", "").replace(".", "").replace(",", "").replace(" ", "").upper()
    print(f"Texto limpio: {texto}\n")
    # Buscamos cadenas repetidas
    cadenas_repetidas = encontrar_subcadenas_repetidas(texto)
    todas_las_distancias = []
    for subcadena in cadenas_repetidas:
        distancias = calcular_distancias_de_subcadenas_repetidas(texto, subcadena)
        todas_las_distancias.extend(distancias)

    return todas_las_distancias

def encontrar_subcadenas_repetidas(texto, min_len=3):
    repetidas = {}
    n = len(texto)
    for length in range(min_len, 8):
        for i in range(n - length + 1):
            sub = texto[i:i+length]
            if sub in repetidas:
                continue
            # Usamos escape para evitar errores con caracteres especiales en regex
            count = len(re.findall(f'(?={re.escape(sub)})', texto))
            if count > 1:
                repetidas[sub] = count
    return repetidas

def calcular_distancias_de_subcadenas_repetidas(texto, subcadena):
    posiciones = [m.start() for m in re.finditer(f'(?={re.escape(subcadena)})', texto)]
    if len(posiciones) < 2:
        return []
    print(f"Subcadena '{subcadena}' encontrada en posiciones: {posiciones}")
    distancias = [posiciones[i+1] - posiciones[i] for i in range(len(posiciones)-1)]
    print(f"Distancias calculadas entre repeticiones: {distancias}\n")
    return distancias

def mostrar_analisis_mcd(distancias):
    if not distancias:
        print("No se encontraron distancias para analizar.")
        return

    print(f"Distancias encontradas: {distancias}")
    mcds = set()
    # Añadimos las distancias mismas como posibles factores
    for d in distancias:
        mcds.add(d)

    for r in range(2, min(len(distancias) + 1, 5)):
        for combo in combinations(distancias, r):
            mcd = math.gcd(*combo)
            if mcd > 1:
                mcds.add(mcd)
    mcds = sorted(list(mcds))
    print(f"Posibles longitudes de clave (MCDs y distancias): {mcds}")
    # Damos el análisis de frecuencias para mostrar las longitudes más probables
    analizar_frecuencia_divisores(distancias)

def analizar_frecuencia_divisores(distancias, max_key_len=20):
    if not distancias:
        return
    conteo_divisores = Counter()
    for d in distancias:
        for i in range(2, max_key_len + 1):
            if d % i == 0:
                conteo_divisores[i] += 1

    print("\nFrecuencia de posibles longitudes de clave:")
    for tam, freq in conteo_divisores.most_common():
        print(f"Longitud {tam}: encontrada como divisor {freq} veces")

def main():
    print("-------- TEST DE KASISKI PARA VIGENERE --------")
    texto = input("Ingrese el texto cifrado: ").replace(" ", "").upper()
    if len(texto) < 3:
        print("Texto demasiado corto.")
        return

    resultado = calcular_cadenas_comunes(texto)
    print(f"\nAnalizando longitudes probables:")
    mostrar_analisis_mcd(resultado)

if __name__ == '__main__':
    main()
```
