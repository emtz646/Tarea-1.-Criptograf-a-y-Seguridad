```python
import unicodedata
from collections import Counter

def descifrar_cesar(texto_cifrado, clave):
    # Alfabeto español con Ñ
    alfabeto = "ABCDEFGHIJKLMNÑOPQRSTUVWXYZ"

    texto_cifrado = texto_cifrado.upper()
    texto_descifrado = ""

    for char in texto_cifrado:
        if char in alfabeto:
            indice_actual = alfabeto.find(char)
            # Módulo 27 para el alfabeto español
            nuevo_indice = (indice_actual - clave) % len(alfabeto)
            texto_descifrado += alfabeto[nuevo_indice]
        else:
            # Mantenemos igual caracteres especiales
            texto_descifrado += char
            continue

    return texto_descifrado

def descifrar_cesar_fuerza(texto_cifrado):
    i = 1
    for clave in range(0, 27):
        resultado = descifrar_cesar(texto_cifrado, clave)
        print(f"INTENTO {i}. Clave {clave}: {resultado}")
        i += 1

def descifrar_cesar_relacion(texto_cifrado):
    alfabeto = "ABCDEFGHIJKLMNÑOPQRSTUVWXYZ"
    relacionA = input("Letra en el texto cifrado: ").upper()
    relacionB = input("Letra que se asume es real: ").upper()

    if relacionA in alfabeto and relacionB in alfabeto:
        indiceA = alfabeto.find(relacionA)
        indiceB = alfabeto.find(relacionB)
        clave = (indiceA - indiceB) % 27
        resultado = descifrar_cesar(texto_cifrado, clave)
        print(f"\nClave calculada: {clave}")
        print(f"Texto descifrado: {resultado}")
    else:
        print("Error: Caracteres de relación no válidos.")

def descifrar_cesar_frecuencias(texto_cifrado):
    alfabeto = "ABCDEFGHIJKLMNÑOPQRSTUVWXYZ"
    # Lista de frecuencia de aparición de letras en el idioma español
    frecuencias_espanol = {
      'E': 13.68,
      'A': 11.82,
      'O': 9.22,
      'S': 7.81,
      'R': 6.86,
      'N': 6.83,
      'I': 6.24,
      'D': 5.86,
      'L': 4.94,
      'T': 4.58,
      'U': 4.33,
      'C': 4.02,
      'M': 3.08,
      'P': 2.89,
      'B': 1.49,
      'G': 1.00,
      'V': 0.90,
      'Y': 0.90,
      'Q': 0.85,
      'H': 0.70,
      'F': 0.69,
      'Z': 0.47,
      'J': 0.44,
      'Ñ': 0.31,
      'X': 0.22,
      'K': 0.11,
      'W': 0.01
    }
    texto_cifrado = texto_cifrado.upper()
    # Tomamos la letra del cifrado que aparece más veces
    letra_mayor_frecuencia = calcular_letra_mayor_frecuencia(texto_cifrado)
    # Comparamos la letra de mayor frecuencia con la lista de las primeras 5 letras con mayor frecuencia en el idioma
    i = 0
    for letra in frecuencias_espanol:
        clave = (alfabeto.find(letra_mayor_frecuencia) - alfabeto.find(letra)) % len(alfabeto)
        resultado = descifrar_cesar(texto_cifrado, clave)
        i += 1
        print(f"INTENTO {i}. Tomando la letra de mayor frecuencia {letra_mayor_frecuencia} como {letra} con frecuencia general de {frecuencias_espanol[letra]}.\n   Clave {clave}: {resultado}")
        if i == 5:
            break

def calcular_letra_mayor_frecuencia(texto):
    texto = texto.upper()
    # Sacamos del análisis a los espacios
    texto_sin_espacios = ''.join(char for char in texto if char.isalpha() and char != ' ' and char in "ABCDEFGHIJKLMNÑOPQRSTUVWXYZ")
    if not texto_sin_espacios:
        return ''
    frecuencias = Counter(texto_sin_espacios)
    letra_mayor = max(frecuencias, key=frecuencias.get)
    return letra_mayor

def main():
    print("-------- DESCIFRADO DE CÉSAR --------")
    texto = input("Ingrese el texto cifrado: ")
    opcion = input("Elija el método de descifrado:\n1 = Fuerza bruta\n2 = Relación\n3 = Análisis de frecuencias\nSelección: ")

    if opcion == "1":
        descifrar_cesar_fuerza(texto)
    elif opcion == "2":
        descifrar_cesar_relacion(texto)
    elif opcion == "3":
        descifrar_cesar_frecuencias(texto)
    else:
        print("Opción no válida")

if __name__ == "__main__":
    main()
```
